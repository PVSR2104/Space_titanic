```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder as o
from sklearn.preprocessing import MinMaxScaler as m
```

```python
data = pd.read_csv('train.csv')
```

```python
data["Age"] = data.Age.fillna(data.Age.mean())
data["HomePlanet"] = data.HomePlanet.fillna("unknown" )
data["CryoSleep"] = data.CryoSleep.fillna(False)
data["Cabin"] = data.Cabin.fillna("unknown" )
data["Destination"] = data.Destination.fillna("unknown" )
data["VIP"] = data.VIP.fillna(False ).replace({True:'True',False:'False'})
```

```python
data[["RoomService",'FoodCourt','ShoppingMall','Spa','VRDeck']] = data[["RoomService",'FoodCourt','ShoppingMall','Spa','VRDeck']].fillna(0,axis=1 )
```

```python
data['Spending'] = data.RoomService+data.FoodCourt+data.ShoppingMall+data.Spa+data.VRDeck
```

```python
#Scalling the data and encoding the data
mscaler = m(feature_range = (0,1))
oscaler = o()
l= ["HomePlanet","CryoSleep","Cabin","Destination","VIP","Transported"]
n = ["RoomService",'FoodCourt','ShoppingMall','Spa','VRDeck',"Spending","Age","Cabin"]
data[l] = oscaler.fit(data[l]).transform(data[l])
data[n] = mscaler.fit(data[n]).transform(data[n])
data[["Cabin","Age"]] = mscaler.fit(data[["Cabin","Age"]]).transform(data[["Cabin","Age"]])
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense,Activation
from keras.optimizers import Adam
from keras.models import Sequential
from keras.metrics import categorical_crossentropy
```

```python
model = Sequential([
    Dense(units = 28,input_shape =(12,),activation = 'relu'),
     Dense(units = 20,activation = 'relu'),
    Dense(units = 10,activation = 'relu'),
     Dense(units = 2,activation = 'Softmax'),
])
```

```python
model.compile(optimizer ='adam',loss = 'sparse_categorical_crossentropy',metrics = ['accuracy'])
```

```python
x = data[["RoomService",'FoodCourt','ShoppingMall','Spa','VRDeck',"Spending","Age","Cabin","HomePlanet","CryoSleep","Destination","VIP"]]
```

```python
y = data["Transported"]
```

```python
from sklearn.utils import shuffle
x,y = shuffle(x,y)
print(x.shape)
print(y.shape)
```

```python
model.fit(x,y,epochs =500,batch_size = 21,validation_split=0.3,shuffle = True)
```
